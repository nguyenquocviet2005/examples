## Check Available Kernels

List all registered Jupyter kernels:

```python
# 2. Check Available Kernels
kernel_dir = Path.home() / '.local' / 'share' / 'jupyter' / 'kernels'

if kernel_dir.exists():
    print(f"ðŸ“ Kernel directory: {kernel_dir}\n")
    kernels = sorted(kernel_dir.glob('*/'))
    print(f"Found {len(kernels)} kernel(s):\n")
    
    for kernel_path in kernels:
        kernel_json = kernel_path / 'kernel.json'
        if kernel_json.exists():
            with open(kernel_json) as f:
                kernel_config = json.load(f)
            
            display_name = kernel_config.get('display_name', 'Unknown')
            language = kernel_config.get('language', 'Unknown')
            print(f"âœ… {kernel_path.name}")
            print(f"   Display Name: {display_name}")
            print(f"   Language: {language}")
            print(f"   Python: {kernel_config['argv'][0]}\n")
else:
    print(f"âš ï¸ Kernel directory not found: {kernel_dir}")
```

## Validate Environment Variables

```python
# 6. Environment Variables
print("ðŸ“‹ Key Environment Variables:\n")

# Get important paths
env_vars = ['PYTHONPATH', 'PATH', 'JUPYTER_PATH', 'JUPYTER_DATA_DIR', 'JUPYTER_CONFIG_DIR', 'VIRTUAL_ENV']

for var in env_vars:
    value = os.environ.get(var)
    if value:
        # Truncate long paths for readability
        if len(value) > 80:
            display_value = value[:40] + "..." + value[-35:]
        else:
            display_value = value
        print(f"{var}:")
        print(f"  {display_value}\n")
    else:
        print(f"{var}: (not set)\n")

# Show venv status
print("Current Virtual Environment:")
print(f"  Executable: {sys.executable}")
print(f"  In venv: {hasattr(sys, 'real_prefix') or (hasattr(sys, 'base_prefix') and sys.base_prefix != sys.prefix)}")
```
